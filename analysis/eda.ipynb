# Exploratory Data Analysis (EDA) - Customer Churn

## Objectives
- Understand the distribution of churn vs. non-churn customers.
- Identify key correlations and drivers of churn.
- Analyze the impact of demographic and account information.
- Explore the relationship between engagement features and churn.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
# Load Data
df = pd.read_csv('../data/processed/final_analytical_dataset.csv')
df.head()
```

## 1. Target Variable Distribution

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Churn', data=df, palette='viridis')
plt.title('Churn Distribution')
plt.show()

print(df['Churn'].value_counts(normalize=True))
```

## 2. Numerical Features Distribution

```python
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges', 'Weekly_Engagement_Score', 'Historical_CLV']

plt.figure(figsize=(15, 10))
for i, col in enumerate(num_cols):
    plt.subplot(2, 3, i+1)
    sns.histplot(data=df, x=col, hue='Churn', kde=True, palette='viridis')
    plt.title(f'{col} Distribution by Churn')
plt.tight_layout()
plt.show()
```

## 3. Categorical Features Analysis

```python
cat_cols = ['Contract', 'PaymentMethod', 'InternetService', 'TechSupport']

plt.figure(figsize=(15, 10))
for i, col in enumerate(cat_cols):
    plt.subplot(2, 2, i+1)
    sns.countplot(data=df, x=col, hue='Churn', palette='viridis')
    plt.title(f'Churn by {col}')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 4. Correlation Matrix

```python
plt.figure(figsize=(12, 10))
# Select numeric columns only
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()
sns.heatmap(corr, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

## 5. Engagement vs Churn

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Churn', y='Weekly_Engagement_Score', data=df, palette='viridis')
plt.title('Weekly Engagement Score vs Churn')
plt.show()
```
